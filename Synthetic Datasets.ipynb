# step 4: Feature Selection:We can also called as variable selection or  attribute selection.Here we will gonna select right features(variables,predictors) for making use of better model.

Feature Selection is a very critical component in Machine Learning  workflow. When presented data with very high dimensionality, models usually choke because
1.Training time increases exponentially with number of features.
2.Models have increasing risk of overfitting with increasing number of features.
Feature Selection methods helps with these problems by reducing the dimensions without much loss of the total information. It also helps to make sense of the features and its importance.

 # step 5: Training and Testing the algorithm:

In this step we will gonna training and testing the data by using algorithm.After analysing,,selecting features we got overall idea about the dataset.so,our aim is to predict the data that is whether the transaction is fraudulant or not .we can easily say that this is a classification problem

```python
# linear model is imported from sklearn
from sklearn import linear_model # linear model is imported from sklearn
```

```python
# from linear_model class we selected the logisticregression algorithm
from sklearn.linear_model import LogisticRegression
```

```python
# train_test_split from model_selection
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(Test_data,Target,test_size=0.35,random_state=42)
```

In above command we are tarining the data and testing the data with respect to dependent variable.

```python
Model=linear_model.LogisticRegression() # applying logistic regression
```

```python
Model1=Model.fit(x_train,y_train) # here fitting function on x_train and y_train.we are fitting the data
```

```python
Yprediction=Model1.predict(x_test) # predicting the x_test
```
